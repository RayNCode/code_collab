This is a second attempt to deal with missing values, using other datasets.

```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
df = pd.read_csv("/content/learn_dataset_job.csv")
```

```python
work_desc = pd.read_csv("/content/code_work_description_map.csv")
```

```python
merged_df = pd.merge(df, work_desc, left_on='work_description', right_on='N3', how='inner')
```

```python
data = pd.read_csv("/content/learn_dataset.csv")
```

```python
data_2 = pd.merge(data, merged_df, on="Id", how="left")
data_2.shape
```

```python
categorical_columns = data_2.select_dtypes(include=['object']).columns
# Remplacer les valeurs manquantes par "None" dans les colonnes catégorielles
data_2[categorical_columns] = data_2[categorical_columns].fillna("None")
# Afficher les premières lignes du DataFrame pour vérifier
data_2.drop(["Id", "target", 'insee_code'], axis="columns")
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

# Sélectionnez les colonnes numériques pour l'imputation
numeric_columns = data_2.select_dtypes(include=['float64', 'int64']).columns

# Créez l'IterativeImputer
imputer = IterativeImputer(estimator=RandomForestRegressor(), max_iter=10, random_state=0)

# Appliquez l'imputation
data_2[numeric_columns] = imputer.fit_transform(data_2[numeric_columns])
```

```python
X = data_2.drop(['N3', 'N2', 'work_description'], axis="columns")
X
```

```python
X['N1'] = X['N1'].round()
X
```

```python
from sklearn.model_selection import train_test_split
import sklearn
sklearn.set_config(display="diagram")
y = data_2['target'].copy()
y = np.where(y == 'B', 1, 0)
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
X.columns
```

```python
preprocessing = make_column_transformer (
    (OneHotEncoder(handle_unknown='ignore'), ["is_student", "OCCUPATION_42", "ACTIVITY_TYPE", "household", "sex", "employer_category", "job_category", "Terms_of_emp",
                       "Eco_sect", "Job_dep", "WORK_CONDITION", "N1"]),
    (OrdinalEncoder(), ["Highest_degree", "EMPLOYEE_COUNT"]),
    remainder='passthrough')
```

```python
logreg = LogisticRegression(solver='lbfgs', max_iter=5000)
pipe_logreg = make_pipeline(preprocessing, logreg)
pipe_logreg
```

```python
cross_val_score(pipe_logreg, X_train, y_train, cv=3, scoring="accuracy").mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(random_state=42)
pipe_forest = make_pipeline(preprocessing, forest_clf)
```

```python
cross_val_score(pipe_forest, X_train, y_train, cv=5, scoring="accuracy", n_jobs=-1).mean()
```

```python
pipe_forest.fit(X_train, y_train)  # Assurez-vous de fournir vos données d'entraînement X_train et y_train
```

```python
# Accédez au modèle RandomForestClassifier à l'intérieur du pipeline
forest_model = pipe_forest.named_steps['randomforestclassifier']

# Obtenez les importances des caractéristiques du modèle
feature_importances = forest_model.feature_importances_
```

```python
# Accédez à l'objet ColumnTransformer dans votre pipeline.
preprocessor = pipe_forest.named_steps['columntransformer']

# Utilisez la méthode get_feature_names_out() pour obtenir les noms des features.
feature_names = preprocessor.get_feature_names_out()

# Reliez ces noms aux importances des caractéristiques pour les interpréter.
feature_importances = pipe_forest.named_steps['randomforestclassifier'].feature_importances_

feature_importances_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

# Trier le DataFrame par importance pour voir les caractéristiques les plus importantes.
feature_importances_df.sort_values(by='Importance', ascending=False)

```
