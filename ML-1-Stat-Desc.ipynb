**First stage, let's see how the variables look like**

```python
df["insee_code"].value_counts()
## Ce sont les codes de communes, on peut voir que Toulouse et Nice (dans une moindre mesure) sont sur représentées.
```

```python
df["AGE_2019"].describe()
## âge de la population étudiée
## les individus sont en moyenne assez âgé. 
```

```python
df["AGE_2019"].hist(bins=50, figsize=(12,6))
## Le dataset est plutôt balancé en termes d'âge. On observe toutefois un net déclin à partir de 70 ans.
## Skewed right : extends much father to the right than to the left.
```
